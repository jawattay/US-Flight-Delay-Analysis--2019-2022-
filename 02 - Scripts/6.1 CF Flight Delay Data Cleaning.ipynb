```python
#identify columns with mixed variable types

def detect_types(series):
    return series.apply(type).nunique()

mixed_type_columns = [col for col in samples_combined.columns if detect_types(samples_combined[col]) > 1]

print("Columns with mixed types:")
print(mixed_type_columns)
```

### The Tail_Number column is the only column with mixed data types. This is to be expected as most tail numbers contain a combination of letters and numbers. I will consider changing this column to an object or string data-type in the future.

```python
#check counts

for column in samples_combined.columns:
    print(f"Counts for column {column}:")
    print(samples_combined[column].value_counts())
    print("\n")
```

```python
#remove duplicates and rename dataframe to more appropriate title

flights_clean = samples_combined.drop_duplicates()
print(flights_clean)
```
